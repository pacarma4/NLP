Boceto del código para la Matriz de Confusión

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

def plot_genre_confusion_matrix(y_true, y_pred, genre_labels):
    """
    Genera y dibuja una matriz de confusión para los géneros musicales.
    
    y_true: Serie o lista con los géneros reales.
    y_pred: Serie o lista con los géneros predichos por el modelo.
    genre_labels: Lista con los nombres de los géneros únicos.
    """
    # 1. Calcular la matriz de confusión
    cm = confusion_matrix(y_true, y_pred, labels=genre_labels) # normalize = 'True'
    
    # 2. Configurar el tamaño del gráfico (importante si hay muchos géneros)
    plt.figure(figsize=(10, 8))
    
    # 3. Dibujar un mapa de calor (heatmap) para que sea visual
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=genre_labels, 
                yticklabels=genre_labels)
    
    # 4. Etiquetas y diseño
    plt.title('Matriz de Confusión: Clasificación de Géneros Musicales', fontsize=16)
    plt.ylabel('Género Real (True)', fontsize=12)
    plt.xlabel('Género Predicho (Predicted)', fontsize=12)
    plt.xticks(rotation=45) # Rotar etiquetas si son largas
    plt.tight_layout()
    plt.show()

# --- EJEMPLO DE USO FUTURO ---
# y_test = df_test['genero_real']
# predicciones = modelo_entrenado.predict(df_test['letras_vectorizadas'])
# generos = df['genero'].unique()
# plot_genre_confusion_matrix(y_test, predicciones, generos)
```

```python
def extraer_top_errores(df, col_real, col_pred, col_letra):
    # Filtramos donde falló el modelo
    errores = df[df[col_real] != df[col_pred]]
    
    # Agrupamos para ver qué pares se confunden más
    top_confusiones = errores.groupby([col_real, col_pred]).size().reset_index(name='frecuencia')
    top_confusiones = top_confusiones.sort_values(by='frecuencia', ascending=False)
    
    return top_confusiones.head(5) # Ver los 5 fallos más comunes
```
